```python
train_neutral = train.query("sentiment=='neutral'")
```

```python
train_neutral.dropna(axis=0)
```

```python
train_neutral['len_text'] = train_neutral['text'].str.len()
train_neutral['len_selected_text'] = train_neutral['selected_text'].str.len()
train_neutral.head(100)
#same_len_df = train_neutral[train_neutral.text.apply(len)==train_neutral.selected_text.apply(len)]
#same_len_df.shape
```

```python
len_tra_neu = len(train_neutral)
same_length = train_neutral[train_neutral.len_text==train_neutral.len_selected_text]
len(same_length)/len_tra_neu*100
```

```python
diff_one_word_len = train_neutral[train_neutral.len_text - 1==train_neutral.len_selected_text]
len(diff_one_word_len)/len_tra_neu*100
```

```python
diff_one_word_len.head()
```

```python
diff_one_word_len = train_neutral[train_neutral.len_text - 2==train_neutral.len_selected_text]
len(diff_one_word_len)/len_tra_neu*100
```

```python
not_same_len = train_neutral[train_neutral.len_text - train_neutral.len_selected_text > 3]
not_same_len.head(100)
```

```python
len(not_same_len)/len_tra_neu*100
```
